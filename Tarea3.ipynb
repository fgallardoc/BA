# Preprocesamiento de los datos 

```python
import pandas as pd
import numpy as np

# Leyendo los datos de un archivo csv:
data = pd.read_csv('X_y_train.csv')
print('Cantidad de Datos (N)= ' + str(len(data)))
print('Cantidad de Variables (M) = ' + str(len(data.columns)))
```

```python
data.info()
```

Borramos el id

```python
data_1 = data.drop('Id', axis=1)
data_1.info()
```

```python
data_1.head()
data_1.info()
```

TRANFORMAMOS LAS VARIABLES CATEGORICAS QUE NO ESTAN COMO ESTE TIPO DE VARIABLE
TRANSFORMAS VARIABLES YES/NO TO 1/0

```python
# falta pasarlo a loop
#Variables binarias ( PASARLO A LOOP DESPUES)
Default = {'yes': 1,'no': 0}
data_1.default = [Default[i] for i in data_1.default]
Loan = {'yes': 1,'no': 0}
data_1.loan = [Loan[i] for i in data_1.loan]
Y = {'yes': 1,'no': 0}
data_1.y = [Y[i] for i in data_1.y]
Housing = {'yes': 1,'no': 0}
data_1.housing = [Housing[i] for i in data_1.housing]
Month = {'jan':1, 'feb':2, 'mar':3, 'apr':4, 'may':5, 'jun':6, 'jul':7, 'aug':8, 'sep':9, 'oct':10, 'nov':11, 'dec':12 } 
data_1.month = [Month[i] for i in data_1.month]
print(data_1) 
```

```python
data_1.head()
```

```python
#loop para transformar variables a categoricas
categoricas = ["job","marital", "education","contact", "month", "poutcome"]
categoricas_1 = ["job_cat","marital_cat", "education_cat","contact_cat", "month_cat", "poutcome_cat"]
for variable in categoricas  :
 data_1[variable] = data_1[variable].astype('category') # reconocimiento como categórica.
 data_1[variable] = data_1[variable].cat.codes
   
print (data_1)
```
